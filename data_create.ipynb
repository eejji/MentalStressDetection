### S2~S17까지 15명의 Normal signal, Stress signal 검출(S11 제외)

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from PIL import *
from tensorflow.keras.preprocessing import image

def max_rpeak(*ecg_v_threshold):
    max=0
    for i in range(1,len(ecg_v_threshold)):
        if ecg_v[ecg_v_threshold[i-1]] < ecg_v[ecg_v_threshold[i]]:
             max = ecg_v_threshold[i]
    return max


person = ['S2','S3','S4','S5','S6','S7','S8','S9','S10','S11','S13','S14','S15','S16','S17']
data_dir = 'C:/Users/jihun/4255/OpensourceHW/WESAD'
save_dir_normal = 'C:/Users/jihun/4255/OpensourceHW/pkl_result/normal'
save_dir_stress = 'C:/Users/jihun/4255/OpensourceHW/pkl_result/stress'
#path = 'C:/Users/jihun/4255/OpensourceHW/WESAD/S5/S5.pkl'
#df = pd.read_pickle(path)

for total in range(len(person)):
    path = data_dir +'/'+ person[total] + '/' + person[total] + '.pkl'
    df = pd.read_pickle(path)
    print(person[total]+'로드완료')
    
    label = df['label']
    
    baseline = np.where(label==1)[0]
    baseline_rpeak=[]
    
    for i in range(len(baseline)):
        if ecg_v[baseline[i]] > 1.3 :
            baseline_rpeak.append(baseline[i])


    ecg_v_threshold = []
    save_number=1
    for count in range(len(baseline_rpeak)):
        if ecg_v[baseline_rpeak[count]] > 1.34:
            ecg_v_threshold.append(baseline_rpeak[count])
            continue
        elif ecg_v[baseline_rpeak[count]] < 1.34:
            ecg_v_threshold.append(baseline_rpeak[count-1])
            print(ecg_v_threshold)
            if len(ecg_v_threshold) < 5:
                    ecg_v_threshold=[]
                    continue
            result = max_rpeak(*ecg_v_threshold)
            print(result)
            plt.figure(figsize=(4,5))
            plt.gca().axes.xaxis.set_visible(False)
            plt.gca().axes.yaxis.set_visible(False)
            plt.plot(np.arange(0,600),ecg_v[(result-300):(result+300)])
            plt.savefig(save_dir_normal+'/'+person[total]+'_'+str(save_number)+'.png')
            print(person[total]+'_'+str(total),count,'저장완료')
            save_number+=1
            plt.show()
            ecg_v_threshold=[]

    #스트레스 
    stress =np.where(label==2)[0]
    stress_rpeak=[]
    
    for i in range(len(stress)):
        if ecg_v[stress[i]] > 1.3 :
            stress_rpeak.append(stress[i])

    ecg_v_threshold = []
    save_number=1
    for count in range(len(stress_rpeak)):
        if ecg_v[stress_rpeak[count]] > 1.34:
            ecg_v_threshold.append(stress_rpeak[count])
            continue
        elif ecg_v[stress_rpeak[count]] < 1.34:
            ecg_v_threshold.append(stress_rpeak[count-1])
            print(ecg_v_threshold)
            if len(ecg_v_threshold) < 5:
                    ecg_v_threshold=[]
                    continue
            result = max_rpeak(*ecg_v_threshold)
            print(result)
            plt.figure(figsize=(4,5))
            plt.gca().axes.xaxis.set_visible(False)
            plt.gca().axes.yaxis.set_visible(False)
            plt.plot(np.arange(0,600),ecg_v[(result-300):(result+300)])
            plt.savefig(save_dir_stress+'/'+person[total]+'_'+str(save_number)+'.png')
            save_number+=1
            plt.show()
            ecg_v_threshold=[]
```
